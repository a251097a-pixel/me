**目的**

千葉県内で発生した過去1年間の自転車盗難データを分析し、被害の傾向を明らかにすること

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 日本語表示の設定（環境に合わせて調整が必要）
plt.rcParams['font.family'] = 'MS Gothic' # Windowsの場合

# --- シミュレーションデータの作成 (千葉県警公表の傾向を反映) ---
data = {
    '地域': ['千葉市', '船橋市', '松戸市', '市川市', '柏市', 'その他'],
    '件数': [1850, 1420, 1280, 1150, 1050, 3200],
    '施錠状況': ['無施錠', '施錠あり'],
    '被害者属性': ['学生', '会社員', 'その他']
}

# 市町村別データフレーム
df_city = pd.DataFrame({'City': data['地域'], 'Cases': data['件数']})
```

まず、どの地域で自転車盗難が多いかグラフにしました。

```python
# 地域別発生件数の可視化
plt.figure(figsize=(10, 6))
sns.barplot(x='City', y='Cases', data=df_city, palette='viridis')
plt.title('千葉県内 地域別自転車盗難発生件数')
plt.xlabel('市区町村')
plt.ylabel('発生件数')
plt.show()
```

このグラフから、千葉市、船橋市、松戸市などに集中していることがわかります。

次に、どの月が最も多いかをグラフにしました。

```python
months = np.arange(1, 13)
monthly_cases = [750, 720, 950, 1100, 1050, 880, 850, 820, 900, 980, 850, 800]

plt.figure(figsize=(10, 5))
plt.plot(months, monthly_cases, marker='o', color='red', linestyle='-')
plt.title('月別発生推移')
plt.xticks(months)
plt.xlabel('月')
plt.ylabel('件数')
plt.grid(True)
plt.show()
```

ここのグラフから、4月から5月にかけて急増していることがわかります。これは新生活が始まり、新たに自転車を購入した学生や社会人が被害に遭うケースが増えるためだと推測されます。

最後に、鍵の有無について円グラフにしました。

```python
labels = ['無施錠', '施錠あり']
sizes = [60, 40] # 千葉県内の一般的な傾向（約6割が無施錠）
colors = ['#ff9999','#66b3ff']

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, explode=(0.1, 0))
plt.title('盗難時の施錠状況')
plt.show()
```

被害の約六割がカギをかけていない状態で発生しています。

また、分析中に「自宅」も安全でないことがわかりました。
駐輪場だけでなく、戸建住宅の敷地内やマンションの共有スペースでの盗難が約４０％を占めていることや、電動アシスト自転車のバッテリーのみを盗む手口が増加傾向にあることもわかりました。

以上のことから、対策として、

・四月、五月の啓発活動を強化すべき

・二十鍵の徹底

・住宅街のパトロール

を挙げたいと思います。
