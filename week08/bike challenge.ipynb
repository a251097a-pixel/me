```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("000067001.csv")
df.head()
df.groupby('発生時（始期）').size().sort_values(ascending=False)
df['時系列'] = pd.to_datetime(df['発生時（始期）'], format="%Y%m%d", errors='coerce')

def categorize_time(hour):
    if 6 <= hour < 9:
        return '朝(通勤通学時間帯)'
    elif 9 <= hour < 17:
        return '昼(日中時間帯)'
    elif 17 <= hour < 23:
        return '夜(帰宅夕方時間帯)'
    elif 23 <= hour < 24 or 0 <=hour < 6:
        return '深夜'
    else:
        return '不明'
df['時系列'] = df['発生時（始期）'].apply(categorize_time)
time_counts = df['Hour'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplpt(x=time_counts.index, y=time_counts.values, palette='plasma')
plt.title('発生時間帯別自転車盗難件数', fontsize=16)
plt.xlabel('発生時間帯', fontsize=12)
plt.ylabel('件数', fontsize=12)
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```
